# Object Oriented Programming

-> instance means the object of a class.  
-> attribute means the values of different variables of an instance or inside some function/method  
--> class atributes --- shared by the all the instances and it belongs to the class.  
--> instance attributes --- specific for the instance and belong to the instance 

```python
class person:
    age=22
    def __init__ (self,name,grade):
        self.name=name
        self.grade=grade
    
    def details(self):
        print(f"the name of the person is {self.name}\n and the grade of the person is {self.grade}\n and the age is {self.age}")
    
    def greet(self):
        print(f"welcome {self.name}....!!")

ob=person("Fawas",12)
ob2=person("usman",11)
ob.details()
ob2.details()
print()
ob.greet()
```

# Inheritance

```python
class animal:
    def __init__(self,sound):
        self.sound=sound
    
    def speak(self):
        print("animal sounds")

    def character(self):
        print("this is the character of the animal")

class dog(animal):
    def speak(self):
        print("woooooolf!!!")   

an=animal("bark")
ab=dog("sounddd")
ab.speak()
ab.character()
```

```python
class animal:
    def __init__(self,sound):
        self.sound=sound
    
    def speak(self):
        print("animal sounds")

    def character(self):
        print("this is the character of the animal")

class dog(animal):
    def speak(self):
        super().speak()
        print("woooooolf!!!")   

an=animal("bark")
ab=dog("sounddd")
ab.speak()
# ab.character()
```

# File Handling

## reading the files

```python
f=open("file_handlng.txt")
print(f.read())
f.close()
```

```python
with open('file_handlng.txt','r') as f:
    print(f.readline())
    print(f.readline())
    print(f.readlines())
```

```python
with open('file_handlng.txt') as f:
    for x in f:
        print(x)
```

## writing and creating the files

```python
with open('file_handlng.txt','a') as f:
    f.write("this is the new line in the file and this is the last line in the file")

with open('file_handlng.txt') as f:
    print(list(x for x in f ))
```
