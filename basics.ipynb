### Python match statments
same as the switch statement of the C++

```python
day=2
match day:
    case 1:
        print("monday")
    case 2:
        print("tuesday")
    case 3:
        print("wednesday")
```

```python
def add_numbers(*args):
    """args (arbitrary arguments) are when we don't know how many values will be passed and args is a tuple"""
    total=0
    for num in args:
        total+=num
    print(f"the total is {total}")
add_numbers(100,101)
help(add_numbers)
```

```python
def details(**kwargs):
    """kwargs (key word arbitrary arguments) is used for dynamic named arguments and its a dictionary"""
    for key,val in kwargs.items():
        print(key,val)
help(details)
details(name="fawas",age=22,college="gpgc")
```

# Decorators
its a special kind of function in python that takes a function as an input and then return another function  
A decorator recieves the function objects not the return values. If we want to recieve the return value we will have to call the function

```python
def change_case(fun):
    def change():
        return fun().upper()
    return change

@change_case
def name():
    return "fawas anayat"
print(name())
```

```python
def combine(f1):
    def decorator(f2):
        def add():
            return f2() +" "+ f1()
        return add
    return decorator

def sec_name():
    return "anayat"

@combine(sec_name)
def first_name():
    return "fawas"

print(first_name())
```

## Lambda Functions
A small anonymous(does not need a name) function that takes any number of arguments and have just one expresssion.  
used for the short operations

```python
fun=lambda x:x*10
print(fun(10))
```

```python
l=[23,45,56,67,88]
a=list(map(lambda num:num*num,l))
print(a)
```

## Recursion
the self calling of the function

```python
def func(n):
    if(n>=0):
        print(n)
        func(n-1)
    else:
        return
func(12)
```

# Exception Handling

```python
x=10
try:  # first this block runs and if it has errors then the except runs and if its the error free then the else runs and at the last finally block runs regardless of what happened in the above all blocks of code
    print(f"the value of x is {x}")
except:
    print("something went wrong")
else:
    print("everything is working fine")
finally:
    print("this is the last block")
```
