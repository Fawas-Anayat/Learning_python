# loc and the iloc
loc ---> select the rows and the columns by their names  ------ df.loc[row_label,col_label]  
iloc ---> select the rows and the columns by their indices.  -----df.iloc[row_index,col_index]

```python
a=df.loc[df['Country']=='Algeria']
```

```python
d=df.iloc[1:45,2:3]
```
