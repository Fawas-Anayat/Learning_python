# view and copy

copy ---> dont update the orirginal array.  
view ---> if changes made in the view then they are also reflected in the original array

```python
arr=np.array([34,45,67,54,423,232])
arrr=arr.copy()
arrr[0]=78213162893
print(arr)
print(arrr)
print(arr.base)
print(arrr.base)
```

```python
arr_1=np.array([34,45,67,54,423,232])
arr_2=arr.view()
arr_2[0]=78213162893
print(arr_1)
print(arr_2)
print(arr_1.shape)
```

## joining different arrays

concatinate is just adding the new array at the end of the first array.  
stack is putting a new array at the new axis .  
hstack function is used to stack across the rows.  
vstack function is used to stack across the columns

```python
arr_6=np.array([3,4,5,56,67,7])
arr_7=np.array([23,34,64,67,6,88])
print(np.concatenate((arr_6,arr_7)))
```

```python
arr_8=np.array([3,4,5,56,67,7])
arr_9=np.array([23,34,64,67,6,88])
print(np.stack((arr_8,arr_9),axis=0))
```

# spliting the arrays
its the opposite of the joining.Means it breaks a single array into the multiple parts

```python
arr_10 = np.array([1, 2, 3, 4, 5, 6])
print(np.array_split(arr_10,2))
```
